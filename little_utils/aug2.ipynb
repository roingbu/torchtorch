```python
import Augmentor

# 1. 指定图片所在目录
#注意此处两个文件夹中图片名必须完全一致，包括后缀名
p = Augmentor.Pipeline(r"casia-face/")  #images中是原图像
# p = Augmentor.Pipeline("./cut128",save_format = "auto")  #images中是原图像

# p.ground_truth('./labels')  #labels中是掩码图像

# 2. 增强操作
# p.resize(probability=1,height=128,width=128)

# 按大小裁剪
p.crop_by_size(probability=1,width=256,height=256,centre=True)

# # 图像旋转，按照0.5的概率，最大左旋转角度20，右旋转角度10
# # 左右的角度必须在25之内
# p.rotate_random_90(probability=0.5)
# # 图像左右、上下、随机左右上下翻转
# p.flip_left_right(probability=0.5)
# p.flip_top_bottom(probability=0.5)
# p.flip_random(probability=1)
# # 旋转90
# p.rotate90(probability=0.5)
# # 图像放大缩小,按照概率为0.5，面积为原来的0.9倍
# p.zoom_random(probability=.5, percentage_area=0.9)
# # 裁剪
# p.crop_random(probability=.5, percentage_area=0.8)
# # 亮度
# p.random_brightness(probability=.2, min_factor=0.1, max_factor=0.9)
# #透视形变-垂直方向形变：magnitude取（0,1），指形变的程度
# p.skew_tilt(probability=0.7,magnitude=0.4)
# #透视形变-斜四角形变形变：magnitude取（0,1），指的是形变程度
# # p.skew_corner(probability=0.7,magnitude=1)

# # 3. 指定增强后图片数目总量
# # 两种操作过后都会在images文件夹下生成output文件夹，生成的结果都存在里面了。

# 第一种生成图像存在重复图像过多和部分图像删减没了的问题，因此我用的一般都是第二种。
#  a.最终扩充的样本数量
p.set_save_format("jpg")
# p.sample(100)
# b.对文件夹中每个图片进行一次操作
p.process()
```

```python
# 如果你需要对每张图片进行一次某操作，就可以这样。
for i in range(1,5):
    if i == 1:
        p.flip_top_bottom(probability=1)
        p.process()
    if i == 2:
        p.flip_top_bottom(probability=1)
        p.flip_left_right(probability=1)
        p.process()
    if i == 3:
        p.flip_left_right(probability=1)
        p.rotate90(probability=1)
        p.process()
    if i == 4:
        p.rotate180(probability=1)
        p.process()
```
